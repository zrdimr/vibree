# Android APK Builder for Google Colab
Run these cells to install the Android SDK, Gradle, and build your APK in the cloud.

## 1. Install Java (JDK 17)

```python
!apt-get update
!apt-get install -y openjdk-17-jdk
import os
os.environ["JAVA_HOME"] = "/usr/lib/jvm/java-17-openjdk-amd64"
```

## 2. Install Android SDK

```python
!mkdir -p android-sdk/cmdline-tools
!wget https://dl.google.com/android/repository/commandlinetools-linux-10406996_latest.zip -O cmdline-tools.zip
!unzip -q cmdline-tools.zip -d android-sdk/cmdline-tools
!mv android-sdk/cmdline-tools/cmdline-tools android-sdk/cmdline-tools/latest

os.environ["ANDROID_HOME"] = "/content/android-sdk"
os.environ["PATH"] += ":/content/android-sdk/cmdline-tools/latest/bin:/content/android-sdk/platform-tools"

!yes | sdkmanager "platform-tools" "platforms;android-34" "build-tools;34.0.0"
```

## 3. Install Gradle

```python
!wget https://services.gradle.org/distributions/gradle-8.5-bin.zip
!mkdir -p /opt/gradle
!unzip -q gradle-8.5-bin.zip -d /opt/gradle
os.environ["PATH"] += ":/opt/gradle/gradle-8.5/bin"
```

## 4. Upload Your Project
**Action Required**: 
1. Zip your local `wearable_data_collector` folder.
2. Upload it to the Files area on the left in Colab (drag and drop).
3. Run the cell below to unzip it.

```python
!unzip -q wearable_data_collector.zip -d project_root
```

## 5. Build the APK

```python
%cd project_root/wearable_data_collector
!chmod +x gradlew
# Use installed gradle if wrapper fails, or just run gradle directly
!gradle :app:assembleDebug
```

## 6. Download APK

```python
from google.colab import files
files.download('app/build/outputs/apk/debug/app-debug.apk')
```
